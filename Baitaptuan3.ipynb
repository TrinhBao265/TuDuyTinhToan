nhap thang trong nam  kiem tra so ngay

```python
thang = int(input())
nam = int(input())

match thang:
    case 1|3|5|7|8|10|12:
        s = f"thang {thang} co 31 ngay"
    case 2:
        if (nam % 4 == 0 and nam % 100 != 0) or (nam % 400 == 0):
            s = f"thang {thang} co 29 ngay"
        else:
            s = f"thang {thang} co 28 ngay"
    case 4|6|9|11 :
        s = f"thang {thang} co 30 ngay"
print(s)        
```
