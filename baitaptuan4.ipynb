[ChickenAndDog] 
"Vừa gà vừa chó, 
Bó lại cho tròn, 
Ba mươi sáu con, 
Một trăm chân chẵn" 
Từ bài toán dân gian trên, mở rộng thành chương trình nhận đầu vào là tổng số
con và tổng số chân của gà và chó. Nếu tìm được số phù hợp, in ra số lượng gà và
W4A15 số lượng chó. Ngược lại, in ra "invalid".

```python
total_animals, total_legs = map(int, input().split())
def chicken_and_dog(total_animals, total_legs):
    if total_legs % 2 != 0:
        print("invalid")
        return
    Cho = (total_legs - 2 * total_animals) // 2
    Ga = total_animals - Cho
    if Cho >= 0 and Ga >= 0 and (Cho * 4 + Ga * 2 == total_legs):
        print(f"Gà: {Ga}, Chó: {Cho}")
    else:
        print("invalid")
print(chicken_and_dog(total_animals, total_legs))


```

[GCD] 
Cho hai số m và n là hai số nguyên dương nhập từ bàn phím. Viết chương trình 
W4A14 tìm ước chung lớn nhất của hai số m và n.

```python
uoc = []
m, n = map(int, input().split())
for i in range(1,n+1):
    if m%i == 0 and n%i == 0:
        uoc.append(i)
print(max(uoc))

        
```
